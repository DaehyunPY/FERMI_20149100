```python
from typing import List, Mapping
from itertools import zip_longest

import numpy as np
from numpy.random import normal, choice, rand
from numpy.polynomial.legendre import Legendre
from scipy.ndimage.filters import gaussian_filter
import matplotlib.pyplot as plt

from vmitools import (
    abel_inverse, finite_legendre_transform_in_theta,
    interp, tohist, transpose_to_drdomega, transpose_to_drdth,
)
```

```python
# Generate dist
def init_dist(beta: List[float], n: int = 1001) -> Mapping[str, np.ndarray]:
    x = np.linspace(-1, 1, n)
    fx = Legendre([1, *beta])(x)
    return {
        'x': x,
        'px': fx / fx.sum(),
    }


beta = [1, 1.5, 0.5, -0.5]
dist = init_dist(beta)


# Generate samples
def init_smp(
    n: int, r: float, sigma: float, x: float, px: float,
) -> Mapping[str, np.ndarray]:
    r = normal(loc=r, scale=sigma, size=n)
    z = r * choice(x, p=px, size=n)
    phi = 2 * np.pi * rand(n)
    x = (r ** 2 - z ** 2) ** 0.5 * np.cos(phi)
    y = (r ** 2 - z ** 2) ** 0.5 * np.sin(phi)
    return {
        'x': x,
        'y': y,
        'z': z,
    }


smp = init_smp(n=1000000, r=8, sigma=0.1, **dist)

# Generate projected image from the samples
zedges = np.linspace(-16, 16, 321)
img, *_ = np.histogram2d(smp['x'], smp['z'], bins=[zedges, zedges])

plt.figure()
plt.pcolormesh(zedges, zedges, img.T)
plt.axis("equal")
plt.show()
```

```python
def mrot(th: float) -> np.ndarray:
    return np.array(
        [[np.cos(th), -np.sin(th)],
         [np.sin(th), np.cos(th)]],
    )


def mhorshear(m: float) -> np.ndarray:
    return np.array(
        [[1, m],
         [0, 1]],
    )


def msqueeze(k: float) -> np.ndarray:
    return np.array(
        [[k, 0],
         [0, 1/k]],
    )



from typing import Callable, Optional
from functools import partial

from numpy import ndarray, sin, cos, abs, pi, meshgrid, stack, array, einsum
from scipy.interpolate import RectBivariateSpline


def transpose_linearly(
    cart: Callable[..., ndarray],
    mat: ndarray,
    x0: Optional[ndarray] = None,
    u0: Optional[ndarray] = None,
) -> Callable[..., ndarray]:
    if u0 is None:
        u0 = array(0)[None]
    if x0 is None:
        x0 = array(0)[None]
    def ret(*u: ndarray) -> ndarray:
        n = len(u)
        shape = slice(None), *(n * [None])
        x = einsum("ij,j...->i...", mat, (stack(u) - u0[shape]))
        return cart(*(x + x0[shape]))
    return ret


f = interp(gaussian_filter(img, 3), zedges, zedges)
img2 = tohist(transpose_linearly(f, msqueeze(1/2)), zedges, zedges)
plt.figure()
plt.pcolormesh(zedges, zedges, img2.T)
plt.axis("equal")
plt.show()

# g = interp(img2, zedges, zedges)
# img3 = tohist(transpose_linearly(g, mrot(-20), x0=np.array([5,5])), zedges, zedges)
# plt.figure()
# plt.pcolormesh(zedges, zedges, img3.T)
# plt.axis("equal")
# plt.show()
```

```python
# Abel invert the image
dz = zedges[1:] - zedges[:-1]
inverted = abel_inverse(img, zedges) * dz[None, :]  # shape: (r, z)

plt.figure()
plt.pcolormesh(zedges, zedges, inverted.T)
plt.axis("equal")
plt.clim(0, 100)
plt.show()
```

```python
# Transpose to other spaces
sliced = interp(inverted, zedges, zedges)  # (rho, z) -> intensity
redges, thedges = np.linspace(0, 16, 801), np.linspace(-np.pi, np.pi, 721)
hist_indrdomega = tohist(transpose_to_drdomega(
    sliced), redges, thedges)  # (r, th) -> intensity
hist_indrdth = tohist(transpose_to_drdth(sliced), redges,
                      thedges)  # (r, th) -> intensity
_, coeff, _ = finite_legendre_transform_in_theta(
    hist_indrdomega.sum(0), thedges, 6,
)  # Shapes of returns: (n, r) (n, r) (n, th)

print("Expected   Simulated")
print("--------   ---------")
for expected, simulated in zip_longest(beta, coeff[1:]/coeff[0]):
    print(f"{expected if expected else 0:>8.3f}   {simulated:>9.3f}")

plt.figure()
plt.hist(np.linalg.norm([smp["x"], smp["y"], smp["z"]], axis=0), redges, histtype="step")
plt.plot((redges[1:] + redges[:-1]) / 2, hist_indrdth.sum(1))
plt.show()
```
